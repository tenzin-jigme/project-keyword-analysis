```python
# Based on competition, replace values for low bid and high bid.
HighComp = df5.loc[df5['competition'] == 'High']
MediumComp = df5.loc[df5['competition'] == 'Medium']
LowComp = df5.loc[df5['competition'] == 'Low']
```

```python
#Calculating Median and assigning variables.
High_median1 = HighComp['low bid'].median()
High_median2 = HighComp['high bid'].median()

Med_median1 = MediumComp['low bid'].median()
Med_median2 = MediumComp['high bid'].median()

Low_median1 = LowComp['low bid'].median()
Low_median2 = LowComp['high bid'].median()


print(High_median1)
print(High_median2)
print()
print(Med_median1)
print(Med_median2)
print()
print(Low_median1)
print(Low_median2)
```

### **Data Imputation.**

```python
#Replacing values.
# High competition keywords
HighComp['low bid'] = HighComp['low bid'].fillna(High_median1)
HighComp['high bid'] = HighComp['high bid'].fillna(High_median2)

# Medium competition keywords
MediumComp['low bid'] = MediumComp['low bid'].fillna(Med_median1)
MediumComp['high bid'] = MediumComp['high bid'].fillna(Med_median2)

# Low competition keywords
LowComp['low bid'] = LowComp['low bid'].fillna(Low_median1)
LowComp['high bid'] = LowComp['high bid'].fillna(Low_median2)
```

```python
# Consolidate the df back into one
pdGroup = [HighComp, MediumComp, LowComp]
cleaned = pd.concat(pdGroup, axis=0)
cleaned.info()
```

## **Data Analysis and Data Visualization**


```python
#Filter data by Average Monthly Searches (AMS).
#Focusing on AMS of 500 (100 to 1,000) and 5000 (1,000 to 10,000).
v1 = cleaned.loc[cleaned['ams'] == 500]
v2 = cleaned.loc[cleaned['ams'] == 5000]
```

# Ad Group 1


```python
v1a = v1.loc[v1['high bid'] > 3]
sortedv1a = v1a.sort_values(by='high bid', ascending=False)
```

```python
# Visualizing the data
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))

sns.barplot(x='high bid', y='keywords', data=sortedv1a, color='#4285f4')

plt.title('Keywords with 500 Avg. Monthly Searches and High Bid above $3', loc='left', fontsize='12', fontweight='bold')
plt.ylabel('')

plt.xlabel('High Bid')
plt.xticks(np.arange(0, 25, step=2))

#plt.savefig('a.png', dpi=300, format='png', bbox_inches='tight')
plt.show()
```

# Ad Group 2


```python
sns.set_style("darkgrid")

fig, axes = plt.subplots(1, 2, figsize=(6,4))

sns.barplot(x = 'low bid', y = 'keywords', data = v2, color='#F4B400', ax = axes[0])
sns.barplot(x = 'high bid', y = 'keywords', data = v2, color='#4285f4', ax = axes[1])
sns.despine

plt.suptitle ("Keywords with 5,000 Avg. Monthly Searches", fontsize=12, fontweight='bold')
axes[0].set_ylabel("")
axes[1].set_ylabel("")
axes[1].set(yticklabels=[])

#plt.savefig('test.png', dpi=300, format='png', bbox_inches='tight')
plt.show()
```
